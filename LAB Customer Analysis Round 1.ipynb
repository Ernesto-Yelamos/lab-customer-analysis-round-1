```python
import pandas as pd
import numpy as np
```

```python
# Read the three files into python as dataframes.
data1 = pd.read_csv("201006_file1.csv")
data2 = pd.read_csv("201006_file2.csv")
data3 = pd.read_csv("201006_file3.csv")
```

```python
# Standardize header names.
data1 = data1.rename(columns = {'ST':'State', 'GENDER':'Gender'})
header1 = list(data1.columns)
data1.head()
```

```python
# change some header names
data2 = data2.rename(columns = {'ST':'State', 'GENDER':'Gender'})
```

```python
# Rearrange the columns in the dataframe as needed.
data2_new = data2[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value', 'Income', 'Monthly Premium Auto', 'Number of Open Complaints', 'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
header2 = list(data2_new.columns)
data2_new.head()
```

```python
# Rearrange the columns in the dataframe as needed.
data3_new = data3[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value', 'Income', 'Monthly Premium Auto', 'Number of Open Complaints', 'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
header3 = list(data3_new.columns)
data3_new.head()
```

```python
# Concatenate the three dataframes.
data = pd.DataFrame(columns = header1)
data = pd.concat([data, data1, data2, data3], axis=0)
```

```python
#Delete the column education and the number of open complaints from the dataframe.
data_copy = data.drop(columns = ['Education', 'Number of Open Complaints'], inplace = False)
data_copy.head()
```

```python
# Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

# delete % sign
data['Customer Lifetime Value'] = data['Customer Lifetime Value'].str.replace('%','').fillna(0)
```

```python
data['Customer Lifetime Value'].isna().sum()
```

```python
# transform to number amd multiply
data['Customer Lifetime Value'] = data['Customer Lifetime Value'].astype('float') * 100
data['Customer Lifetime Value']
```

```python
#Check for duplicate rows in the data and remove if any.
data.drop_duplicates()
```

```python
# Filter out the data for customers who have an income of 0 or less.
data = data[(data['Income'] <= 0)]
data
```
